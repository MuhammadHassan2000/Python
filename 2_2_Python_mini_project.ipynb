```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris
import pandas as pd
```

```python

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
file_path = '/iris_dataset.csv'
df_iris.to_csv(file_path, index=False)
```

**Matplotlib**

```python
plt.figure(figsize=(8, 6))
for species in df_iris['species'].unique():
    subset = df_iris[df_iris['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)
plt.title('Matplotlib: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

**Seaborn**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', style='species', data=df_iris, palette="deep", s=100)
plt.title('Seaborn: Sepal Length vs Sepal Width')
plt.show()
```

**Plotly** **Express**

```python
fig = px.scatter(df_iris, x='sepal length (cm)', y='sepal width (cm)', color='species',
                 title='Plotly Express: Sepal Length vs Sepal Width', labels={'sepal length (cm)': 'Sepal Length (cm)', 'sepal width (cm)': 'Sepal Width (cm)'})
fig.show()
```
